```python
import os 
import numpy as np
import matplotlib.pyplot as plt
from fxpmath import Fxp
```

```python
n_word = 16
n_frac = 14
IMAGE_SIZE = (50, 80)
NUM_BLOBS = 1
CROP_BOX_ROWS = 20
CROP_BOX_COLS = 20
```

```python
tb_image = np.zeros(IMAGE_SIZE[0]*IMAGE_SIZE[1]).astype(int)
dest_idx = 0
for i in range(IMAGE_SIZE[0]):
    for j in range(IMAGE_SIZE[1]):
        tb_image[dest_idx] = dest_idx
        dest_idx += 1
```

```python
tb_image_fxp = Fxp(tb_image, signed=True, n_word=n_word, n_frac=0)
fpath_tb_image = os.path.join("tb_data", f"tb_image_INDEX_{IMAGE_SIZE[0]}x{IMAGE_SIZE[1]}_{NUM_BLOBS}crops_{CROP_BOX_ROWS}x{CROP_BOX_COLS}_ap_fixed_{n_word}_{n_word-n_frac}.bin")

with open(fpath_tb_image, "wb") as f:
    for i in range(len(tb_image_fxp)):
        str_to_write = tb_image_fxp[i].bin() + "\n"
        f.write(str_to_write.encode())
```
